`````python
def generate_connl(sentence):
    system_prompt = {
        "role":"system",
        "content":"You are data preparation tool for connl format --- You are given text in uzbek language your task is to convert it to connl data with entities annoted format. Output in json format. Json schema {'text':'', 'entities':[]}"
    }
    sentence = {
        "role":"user",
        "content": sentence,
    }
    output = openai_client.chat.completions.create(
        model="gpt-4-0125-preview",
        messages=[
            system_prompt, 
            sentence
        ]
        
    )
    print(output.usage)
    return output.choices[0].message.content.replace("\n", "")
`````

`````python
schema_1  = {
    "type": "array",
    "items": {
        "type": "object",
        "properties": {
            "sentence": {
                "type": "string"
            },
            "entities": {
                "type": "array",
                "items": {
                    "type": "object",
                    "properties": {
                        "start": {
                            "type": "integer"
                        },
                        "end": {
                            "type": "integer"
                        },
                        "label": {
                            "type": "string",
                            "enum": ["PERSON", "ORG", "LOC", "MISC"]  # Assuming possible entity labels
                        }
                    },
                    "required": ["start", "end", "label"]
                }
            }
        },
        "required": ["sentence", "entities"]
    }
}
`````

`````python
import json

def convert_json_to_conll(json_file, output_file):
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    with open(output_file, 'w', encoding='utf-8') as f:
        for item in data:
            sentence = item['sentence']
            entities = item['entities']
            
            tokens = sentence.split()
            token_to_tag = ['O'] * len(tokens)
            
            for entity in entities:
                start_char = entity['start']
                end_char = entity['end']
                label = entity['label']
                
                start_token = len(sentence[:start_char].split())
                end_token = len(sentence[:end_char].split())
                
                if start_token < len(tokens):
                    token_to_tag[start_token] = f'B-{label}'
                if end_token < len(tokens) and start_token != end_token:
                    for i in range(start_token + 1, end_token + 1):
                        token_to_tag[i] = f'I-{label}'
            
            for token, tag in zip(tokens, token_to_tag):
                f.write(f'{token} {tag}\n')
            f.write('\n')

# Usage
convert_json_to_conll('ttt.json', 'data_2.txt')
`````
